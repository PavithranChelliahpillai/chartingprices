```python
import yfinance as yf
import matplotlib.pyplot as plt, mpld3
import time
from datetime import datetime
from pytz import timezone
import pandas as pd 
import sys
import os
from matplotlib.animation import FuncAnimation

def coloured(r, g, b, text):
    return "\033[38;2;{};{};{}m{} \033[38;2;255;255;255m".format(r, g, b, text)
    
yf.pdr_override()

table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]

snp500list = []
for symbol in df['Symbol']:
  snp500list.append(symbol)

#choice=input("Enter ticker: ")
#choice=choice.upper()

choice="BTC-USD"

fmt = "%H:%M:%S"
now_time = datetime.now(timezone('US/Eastern'))

def get_current_price(symbol):
    ticker = yf.Ticker(symbol)
    todays_data = ticker.history(period='1d',rounding=True)
    return todays_data['Close'][0]

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("dark_background")
for param in ['text.color', 'axes.labelcolor', 'xtick.color', 'ytick.color']:
    plt.rcParams[param] = '#212953' 
for param in ['figure.facecolor', 'axes.facecolor', 'savefig.facecolor']:
    plt.rcParams[param] = '#3b0400'
colors = [ 
    '#39FF14' 
]
maxmin=[]
#start = datetime.time(9, 30, 0)
#end = datetime.time(16, 30, 0)
a = True
b=0
tickerlist=[]
while a:
  #currTime = datetime.datetime.now().time()
  #if (currTime>=start and currTime<=end):
    #a=True
  while get_current_price(choice)==b:
      get_current_price(choice)
  z=get_current_price(choice)
  tickerlist.append(z)
  maxmin.append(z)
  maxmin.sort()
  maxo=maxmin[0]*1.005
  mino=maxmin[-1]*0.995
  b=z
  df = pd.DataFrame(tickerlist)

  fig, ax = plt.subplots()

  df.plot(marker='', color=colors, ax=ax)
  n_shades = 10
  diff_linewidth = 1.55
  alpha_value = 0.3 / n_shades
  for n in range(1, n_shades+1):
    df.plot(marker='',
            linewidth=2+(diff_linewidth*n),
            alpha=alpha_value,
            legend=False,
            ax=ax,
            color=colors)
  for column, color in zip(df, colors):
    ax.fill_between(x=df.index,
                    y1=df[column].values,
                    y2=[0] * len(df),
                    color=color,
                    alpha=0.1)
  ax.set_xticks([])
  ax.set_yticks([])
  ax.grid(False)
  ax.axis('off')
  ax.legend().set_visible(False)
  ax.set_xlim([0,100])
  ax.set_ylim([mino,maxo])
  now = datetime.now()
  time.sleep(30)
  print (now.strftime(fmt))
  plt.show()
  html_graph = mpld3.fig_to_html(fig)

```
